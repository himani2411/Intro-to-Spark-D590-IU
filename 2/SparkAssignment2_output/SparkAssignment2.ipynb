Connecting to Spark Cluster

```python
from pyspark import SparkContext
sc = SparkContext()
```

```python
# import nltk
```

Specifying the Location of file and creating a RDD from README.md file

```python
dataFile = "file:///Users/himanideshpande/Downloads/assignment_2_datafile.txt"
```

```python
data = sc.textFile(dataFile).cache()
```

Splitting the data file

```python
splitdata = data.flatMap(lambda line: line.split(" "))#.map(lambda x: (x, len(x)))
```

```python
splitdata.collect()
```

Filtering the data to keep all the words which are above 3 in length and keeping words in lower case.

```python
filtersplitdata = splitdata.map(lambda x : (x.lower(), len(x))).filter(lambda t : t[1]>3)
```

```python
filtersplitdata.collect()
```

Creating a RDD which has key-value of words and their counts in the file

```python
countdata = filtersplitdata.map(lambda x : (x[0], 1)).reduceByKey(lambda a,b: a+b)
```

```python
countdata.collect()
```

```python
countdata.saveAsTextFile("IPYNB/CountDataOutput.txt")
```
